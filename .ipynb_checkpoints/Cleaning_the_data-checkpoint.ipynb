```python
# Read the data file and take a look at the data
local_folder = '.'
df = pd.read_sas(os.path.join(local_folder, 'hints5_cycle3_public.sas7bdat'),encoding='mac_roman')

#display all columns and the first couple of rows
pd.set_option('display.max_columns', None)
df.head(15)
```

Next we begin cleaning the data. 

Survey data is prone to missing data in the form of non-responses and/or indeterminate data. In the HINTS data set, the missing information is identified using the following values:

**-1**: Valid skips or appropriately missing data following a dependent question (correctly skipped)

**-2**: The question was answered in error by the respondent

**-4**: unreadable or non-conforming numeric response)

**-5**: Respondent selected more response options than appropriate for the question

**-6**: Missing data in variables following a missing filter question

**-7**: Missing data (exclusive to web mode respondent variables)

**-9**: Missing data. Not ascertained

We need to figure out if the missing data is randomly missing or not. We can do this by visual inspect as done below for the Age variable


```python
df.loc[df.Age == -9.0,:].head(60)
```

It appears the missing Age data is randomly missing. A similar inspections on other variables of interest yields similar results.
Suppose for interest's sake that we remove the rows with missing Age data (this includes rows with -9, -7 and -4 values). The result dataframe will be as follows

```python
#replace missing data with NaN, remove Age values with NaN
df['Age'].replace(-9.0, np.nan, inplace = True)
df['Age'].replace(-4.0, np.nan, inplace = True)
df['Age'].replace(-7.0, np.nan, inplace = True)

drop_df=df.dropna(subset=["Age"])
```

We have many other variable of interest in this dataset, including some (yes/no) type questions eg.:**'SeekHealthInfo'**, **'UseInternet'**,**'Electronic_SelfHealthInfo'**, **'Electronic_BuyMedicine'**,**'Electronic_TalkDoctor'**,**'Electronic_TrackedHealthCosts'**,**'Electronic_MadeAppts'** and others. A better method for filling the missing data in these columns might be imputation. We do this below for the columns of interest. It is still necessary to convert the values to NaN first.


```python
#convert missing data with NaN before imputing the data
df['SeekHealthInfo'].replace(-9.0, np.nan, inplace = True)
df['UseInternet'].replace(-9.0, np.nan, inplace = True)
df['Electronic_SelfHealthInfo'].replace(-9.0, np.nan, inplace = True)
df['Electronic_BuyMedicine'].replace(-9.0, np.nan, inplace = True)
df['Electronic_TalkDoctor'].replace(-9.0, np.nan, inplace = True)
df['Electronic_TrackedHealthCosts'].replace(-9.0, np.nan, inplace = True)
df['Electronic_MadeAppts'].replace(-9.0, np.nan, inplace = True)

cols_to_impute = ['SeekHealthInfo', 'UseInternet','Electronic_SelfHealthInfo', 'Electronic_BuyMedicine','Electronic_TalkDoctor','Electronic_TrackedHealthCosts','Electronic_MadeAppts']
#cols_to_impute = ['SeekHealthInfo', 'UseInternet']
#cols_to_impute = ['SeekHealthInfo', 'UseInternet']

#imputation 
imputer = IterativeImputer()
imputed_df = pd.DataFrame(imputer.fit_transform(df[cols_to_impute]))
imputed_df.columns = cols_to_impute
drop_df.loc[:,cols_to_impute] = imputed_df.loc[:,cols_to_impute]

#drop_df
```

Now that we have filled data in the columns of interest we can begin to investigate trends and correlations in the data. We will use a clustering algorithm to determine if there are any similarities between groups that answer the above mentioned questions the same way. We will use ***K-Means clustering*** in this case. This analysis is based on [this](https://medium.com/@blazetamareborn/practicing-clustering-techniques-on-survey-dataset-f7d7a322e6ff) article

```python
# reduce to 2 dimensions
pca = PCA(n_components = 2, random_state=1)
Imp_pca = pca.fit_transform(imputed_df)


#Find the best k
distortions = []
K_to_try = range(1, 6)

for i in K_to_try:
    model = KMeans(
            n_clusters=i,
            init='k-means++',
 #           n_jobs=-1,
            random_state=1)
    model.fit(Imp_pca)
    distortions.append(model.inertia_)
plt.plot(K_to_try, distortions, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion')
plt.show()
```

```python
# use the best K from elbow method
model = KMeans(
    n_clusters=4,
    init='k-means++',
#    n_jobs=-1,
    random_state=1)

model = model.fit(Imp_pca)
y = model.predict(Imp_pca)

#plot the clusters

plt.scatter(Imp_pca[y == 0, 0], Imp_pca[y == 0, 1], s = 50, c = 'yellow', label = 'Cluster 1')
plt.scatter(Imp_pca[y == 1, 0], Imp_pca[y == 1, 1], s = 50, c = 'green', label = 'Cluster 2')
plt.scatter(Imp_pca[y == 2, 0], Imp_pca[y == 2, 1], s = 50, c = 'red', label = 'Cluster 3')
plt.scatter(Imp_pca[y == 3, 0], Imp_pca[y == 3, 1], s = 50, c = 'black', label = 'Cluster 4')


plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s = 100, c = 'blue', label = 'Centroids')
plt.title('Clusters of S')
plt.xlabel('Principle Component 1')
plt.ylabel('Principle Component 2')
plt.legend()
plt.grid()
plt.show()

```

Now we use the clustering data to learn about more about the respondents. Let us investigate the spread of age in the clusters


```python
#imputed_df["Age"]=drop_df["Age"]
```

```python
y = pd.DataFrame(y, columns=['cluster'])
clust_df = pd.concat([drop_df.reset_index(), y], axis=1)

#group the data point from same cluster number __
group_clust=clust_df.loc[clust_df.cluster==0,:]

sns.boxplot(group_clust.Age)
plt.show()
```
